Now lets see how to concat

```python
india=pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
india
```

```python
us=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
    "humidity":[68,65,75]
})
us
```

```python
df1=pd.concat([india,us])
df1
```

```python
df2=pd.concat([india,us],ignore_index=True)
df2
```

```python
temp_df=pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30]
})
temp_df
```

```python
humidity_df=pd.DataFrame({
    "city":["mumbai","delhi","agra"],
    "humidity":[80,60,78]
})
humidity_df
```

```python
df3=pd.concat([temp_df,humidity_df],axis=1)
df3
```

now, intelligently add

```python
df4=pd.merge(temp_df,humidity_df,on="city",how="inner")
df4
```

```python
df5=pd.merge(temp_df,humidity_df,on="city",how="outer")
df5
```

```python
df6=pd.merge(temp_df,humidity_df,on="city",how="left")
df6
```

```python
df7=pd.merge(temp_df,humidity_df,on="city",how="right")
df7
```

```python
df8=pd.merge(temp_df,humidity_df,on="city",how="outer",indicator=True)
df8
```

concat--> column and dataframe

```python
s=pd.Series(["rain","sunny","windy"],name="event")
```

```python
df9=pd.concat([temp_df,s],axis=1)
df9
```

```python
df10=pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
df10
```

```python
df11=pd.DataFrame({
    "city":["mumbai","delhi","bangalore"],
    "temperature":[22,55,20],
    "humidity":[50,70,38]
})
df11
```
